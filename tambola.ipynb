```python
import random

def func():
    ticket = random.sample(range(1, 21), 5)
    ticket.sort()
    return ticket

n = int(input("\nEnter no of players: "))
tkt = {}

for i in range(1, n + 1):
    tkt[i] = func()

x = set()
print("\nGame Start ")
print("Rule:Players will remove their tickets as per announcement")
while True:
    num = int(input("\nCall a number btwn 1-20: "))
    if num < 1 or num > 20:
        print("Invalid no Enter between 1 and 20.")
        continue

    if num in x:
        print("Number already called")
        continue

    x.add(num)

    for player in tkt:
        if num in tkt[player]:
            tkt[player].remove(num)
            print("Number ",num," removed from Player ",player,"'s ticket!")

            if len(tkt[player]) == 0:
                print("\nPlayer no:",player," WINS")
                print("\nRemaining tickets: ")
                for i in tkt:
                    if i != player:
                        print("Player ",i,":",tkt[i])
                exit()
```
